```python
data_lst = []
# Read the file.
with open('submission_complete.csv', 'r', encoding="ISO-8859-1") as f:
    line_cnt = 1
    for line in f:
        sentence = line.split("|")  # Split the line with "|".
        temp = []
        for i in range(len(sentence)):
            s = sentence[i].strip()
            temp.append(s)
        data_lst.append(temp)  # Add to data_lst.
    line_cnt += 1
```
