```python
# Inputs of all data.
key = input()  # Key word.

# Store inputs as a sting with empty space.
text_total = ""
while True:
    line = input().strip()
    if line == "INPUT_END":
        # Condotion to stop this program.
        break
    else:
        text_total = text_total + line + " "


# Build find function to find key word.
def find(text, number, key, answer_list):
    # Number of key word appearing in the text after index(number).
    times = text.find(key, number)
    if times >= 0:
        # Situation when key word appears in the text after index(number).
        index = text.find(key, number)
        if index < 7:
            # Situation when the key word to the beginning is smaller than 7.
            # Then build new sentence as required with "**" sign.
            sentence_one = text[:index]
            sentence_one = sentence_one + "**" + key + "**"
            sentence_two = text[index+len(key):index+len(key)+7]
            target = sentence_one + sentence_two
        elif len(text) - index - 1 <= 7:
            # Situation when the key word to the end is smaller than 7.
            # Then build new sentence as required with "**" sign.
            sentence_one = text[index-7:index]
            sentence_one = sentence_one + "**" + key + "**"
            sentence_two = text[index+len(key):]
            target = sentence_one + sentence_two
        else:
            # Other situation.
            # Then build new sentence as required with "**" sign.
            sentence_one = text[index-7:index]
            sentence_one = sentence_one + "**" + key + "**"
            sentence_two = text[index+len(key):index+len(key)+7]
            target = sentence_one + sentence_two
        # Store target in the answer_list.
        answer_list.append(target)
        times -= 1
        number = index + 1
        # Check the rest sentence after key word next time.
        find(text, number, key, answer_list)
    else:
        return

# Store answer in answer_list.
answer_list = []
start_number = 0

# Use find function to find key word of inputs.
find(text_total, start_number, key, answer_list)

# Print out results.
if len(answer_list) == 0:
    # Situation when no sentence matched key word.
    print("NO_MATCH")
else:
    # Situation when sentences match key word.
    for i in range(len(answer_list)):
        print(answer_list[i].rstrip())
```
