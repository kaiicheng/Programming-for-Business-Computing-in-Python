```python
# Inputs of all data.
gap = int(input())  # Distance between key words assigned.
key_one = input()  # Key word one.
key_two = input()  # Key word two.

# Store inputs as a sting with empty space.
text = ""
while True:
    line = input().strip()
    if line == "INPUT_END":
        # Condotion to stop this program.
        break
    else:
        text = text + line + " "


# Build find_two function to find key word two.
def find_two(text, target, index_one, key_one, key_two, answer_list):
    temp_num = target.find("**") + 2
    # Distance from the beginning of target to key word one.
    shift_one = target.find(key_one, temp_num)
    # Reference number to check index of text.
    ref_text = index_one
    # Reference number to check index of taret.
    ref_target = shift_one
    for i in range(len(target)):
        # Distance from key word one to key word two in target.
        shift_two = target.find(key_two, ref_target)
        # Find the index of key_two after key_one in the target string.
        index_two = text.find(key_two, ref_text)
        if shift_two >= 0:
            # Situation when key word two appearing after key word one.
            index_one = index_one  # Index of key_one in text string.

            # Build the sentence as required.
            part_one = target[:shift_one]
            part_one = part_one + key_one + "**"
            part_two = text[index_one + len(key_one):index_two]
            part_three = "**" + key_two + "**"
            temp_number = index_two + len(key_two)
            part_four = text[temp_number: temp_number + 7]
            answer = part_one+part_two+part_three+part_four

            # Store answer in the answer_list.
            answer_list.append(answer)
            ref_text = index_two + 1
            ref_target = shift_two + 1
            # Check the rest sentence after key word next time.
        else:
            # Situation when no key word two appearing after key word one.
            return


# Build find function to find key word one.
def find(text, number, key_one, key_two, answer_list, gap):
    # Calculate the times of key word one appearing in text.
    times = text.find(key_one, number)
    if times >= 0:
        # Situation when key word one appears in text.
        index = text.find(key_one, number)
        if index - gap - len(key_two) < 0:
            # Situation when the key word to the beginning is smaller than 7.
            # Then build new sentence as required with "**" sign.
            sentence_one = text[:index]
            sentence_one = sentence_one + "**" + key_one + "**"
            sentence_two = text[index+len(key_one):index +
                                (len(key_one)+len(key_two)+gap)]
            target = sentence_one + sentence_two
        elif len(text) - (index + len(key_one) + len(key_two) + 1) <= 0:
            # Situation when the key word to the end is smaller than 7.
            # Then build new sentence as required with "**" sign.
            sentence_one = text[index-len(index)+6:index]
            sentence_one = sentence_one + "**" + key_one + "**"
            sentence_two = text[index+len(key_one):]
            target = sentence_one + sentence_two
        else:
            # Other situation.
            # Then build new sentence as required with "**" sign.
            sentence_one = text[index-7:index]
            sentence_one = sentence_one + "**" + key_one + "**"
            sentence_two = text[index+len(key_one):index +
                                (len(key_one)+gap+len(key_two))]
            target = sentence_one + sentence_two

        # Then use find_two function to find key word two.
        number_one = index
        find_two(text, target, number_one, key_one, key_two, answer_list)

        times -= 1
        index_one = index + 1
        # Check the rest sentence after key word next time.
        find(text, index_one, key_one, key_two, answer_list, gap)
    else:
        # Situation when no key word one appears in text.
        return

# Store answer in answer_list.
answer_list = []
number = 0

# Use find function to find key word of inputs.
find(text, number, key_one, key_two, answer_list, gap)

# Print out results.
if len(answer_list) == 0:
    # Situation when no sentence matched key word.
    print("NO_MATCH")
else:
    # Situation when sentences match key word.
    for i in range(len(answer_list)):
        print(answer_list[i].rstrip())
```
